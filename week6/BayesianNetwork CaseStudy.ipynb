```python
# Install pgmpy (run once)
!pip install pgmpy

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination
import pandas as pd

# Create dataset
data = pd.DataFrame({
    'study': ['yes', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'no'],
    'IQ':    ['high', 'low', 'high', 'low', 'high', 'low', 'high', 'low'],
    'pass':  ['yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no']
})

# Define Bayesian Network structure
model = DiscreteBayesianNetwork([
    ('study', 'pass'),
    ('IQ', 'pass')
])

# Train the model
model.fit(data)

# Print CPDs
for cpd in model.get_cpds():
    print(cpd)

# Perform inference
inference = VariableElimination(model)

result = inference.query(variables=['pass'],
                         evidence={'study': 'yes'})

print(result)
```
